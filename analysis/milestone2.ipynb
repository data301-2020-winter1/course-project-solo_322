```python
plt.rc("axes.spines", top=False, right=False)
plt.figure(figsize=(7,10))
violin_plot = sns.violinplot(x = 'educ', y = 'ppage', hue = "gender", data = df1, split = True).set(xlabel = 'Education', ylabel = 'Age')
```

```python
#DataFrame 1
#All incorrect or missing values dropped

#The average age of the respondants is about 52 which signifies that the majority of the population is adult

#No missing values have been reported

#The plot shows the distribution of age vs the education level of the respondents (with an additional attribute of gender)

#It should that there is not much difference in education in terms of gender

#Younger people probably tend to be more qualified comparative to the older ones
```
